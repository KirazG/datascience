```python
import matplotlib.pyplot as plt
import pandas as pd
# Data to plot

df = pd.read_excel('data/case-status.xls',header=1)
dfyear = df.loc[df['Year'] == 2017]
accepted_total = dfyear['Accepted'].sum()
rejected_total = dfyear['Rejected'].sum()
labels = ['Accepted','Rejected']
sizes = [accepted_total,rejected_total]
colors = ['green', 'red','yellow','red','purple','orange']
explode = (0.1, 0.1)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```
